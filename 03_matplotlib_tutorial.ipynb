```python
df = pd.read_csv('Iris.csv', index_col = 0)
df.shape
```

```python
df.head()
```

```python
plt.scatter(x=df['SepalLengthCm'],y=df['SepalWidthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('iris')
plt.show()
```

```python
df['Species'].unique()
```

```python
df['Species'].value_counts()
```

```python
seto = df[df['Species']=='Iris-setosa']
versi = df[df['Species']=='Iris-versicolor']
virgin =  df[df['Species']=='Iris-virginica']

plt.scatter(x=seto['SepalLengthCm'],y=seto['SepalWidthCm'], color='r', label = 'setosa')
plt.scatter(x=versi['SepalLengthCm'],y=versi['SepalWidthCm'], color='g', label = 'versicolor')
plt.scatter(x=virgin['SepalLengthCm'],y=virgin['SepalWidthCm'], color='b', label = 'virginica')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('Iris')
plt.legend()
plt.show()
```

```python
seto = df[df['Species']=='Iris-setosa']
versi = df[df['Species']=='Iris-versicolor']
virgin =  df[df['Species']=='Iris-virginica']

plt.scatter(x=seto['PetalLengthCm'],y=seto['PetalWidthCm'], color='r', label = 'setosa')
plt.scatter(x=versi['PetalLengthCm'],y=versi['PetalWidthCm'], color='g', label = 'versicolor')
plt.scatter(x=virgin['PetalLengthCm'],y=virgin['PetalWidthCm'], color='b', label = 'virginica')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('Iris')
plt.legend()
plt.show()
```

```python
mydata = [[1,2,3,4],
          [10,11,12,13],
          [30,31,32,33],
          [41,42,45,45]]
fig, ax = plt.subplots(2,2)

ax = ax.reshape(-1)
for i in range(4):
    ax[i].plot(mydata[i])
```

```python
img = Image.open('dog.jpg')
arr = np.array(img)
plt.figure(figsize=(5,5))
plt.imshow(arr)
plt.xticks([]) # x 좌표 눈금 없애주기
plt.yticks([]) # y 좌표 눈금 없애주기
plt.title('DOG')
```

## 이미지 여러개

```python
mydata = [np.array(Image.open('dog.jpg')),np.array(Image.open('dog2.jpg')),np.array(Image.open('dog3.jpg'))]
title = ['Dog1','Dog2','Dog3']
fig, ax = plt.subplots(1,3, figsize = (10,5))

for x,z,t in zip(ax,mydata,title):
    x.imshow(z)
    x.set_xticks([])
    x.set_yticks([])
    x.title.set_text(t)
```
